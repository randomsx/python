## 单变量分析绘图
查看数据分布 通常做单变量单特征的分析

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats,integrate

sns.set(color_codes=True)
np.random.seed(sum(map(ord,"distributions")))
x=np.random.normal(size=100)
sns.distplot(x,kde=False) # 直接调用sns.displot 绘制直方图 # 只需要传入数据即可 自动划分横纵坐标和每个条的大小
#kde 是否绘制高斯核密度估计。
```

```python
sns.distplot(x,bins=20,kde=False) # bins 表示把当前数据切分为几大块 这里设置成切分为20小块
```

看数据整体分布的状况和轮廓
displot 绘制直方图

```python
x=np.random.gamma(6,size=200)
sns.distplot(x,kde=False,fit=stats.gamma) # 可以将fit写进来 就是上面的曲线
```

## 特征和特征之间的关系  用散点图 是最好的办法进行描述

```python
# 生成随机数据
mean,cov=[0,1],[(1,.5),(.5,1)] # 指定数据的均值和协方差
data=np.random.multivariate_normal(mean,cov,200) # 使用该函数生成200个数据
df=pd.DataFrame(data,columns=["x","y"])
df
```

joinplot 函数 轻松绘制散点图  一方面绘制出散点图 另一方面把单变量的直方图也绘制出来了
* 双变量关系图jointplot

```python
sns.jointplot(x='x',y='y',data=df); # 将两个变量传进去
```

hex 图 对于比较多的数据 可以更好地观测 通过颜色的差异 分辨出哪块的密度更大

```python
x,y=np.random.multivariate_normal(mean,cov,1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x,y=y,kind='hex',color='k')
```

## 盒图
数据集当中可能有些离群点 
* IQR即统计学概念四分之距，第一/四分位与第三/四分位之间的距离 即数量之间的差
* N=1.5IQR 如果一个值>Q3+N <Q1-N,则为异常点

两端是最大值和最小值  中间的横杠 下面是四分之一位 上面是四分之三位， 上面的菱形就是离群点

```python
sns.boxplot(x='day',y='total_bill',hue='time',data=tips)
```

```python
sns.violinplot(x='total_bill',y='day',hue='time',data=tips) # 小提琴图
```

## 分类属性绘图

* barplot 条形图
* pointplot 点图  看各自的变化 点图可以更好地描述差异性
* factplot 多层面板分类图 比较常用 默认是折线图

```python
sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips)  
```

```python
sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips,kind='bar')
```
