# matplotlib画图

```python
import matplotlib as plt
data=pd.read_csv('data.csv',usecols=['timestamp','label1'])
data.head(2)
```

# 子图绘制 subplot

```python
import matplotlib.pyplot as plt
fig=plt.figure() # 创建图表区域
ax1=fig.add_subplot(4,3,1) #  前两个参数是图像figure尺寸  最后一个参数表示在这个画布中的位置
ax2=fig.add_subplot(4,3,2)
ax3=fig.add_subplot(4,3,6)
plt.show()
```

```python
import numpy as np  
fig=plt.figure(figsize=(3,3))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))
plt.show()
```

# 绘制条形图和散点图
确定条形图的高度和离x轴的距离

```python
data=pd.read_csv('./all/data.csv')
```

```python
from numpy import arange
num_cols=['label1','label2'] # 拿出要画图 即要计算的列
num_reviews=data[num_cols] # 取得特定列
bar_heights=num_reviews.loc[0,num_cols].values.astype('float') # ix 选取单个列或列子集 ix[x,y] 第x行y列
print(bar_heights)

bar_positions=arange(2)+0.75 # arange(5) 0 到5 以1 为步长 加0.75  这样可以实现 不同条不同横坐标
print(bar_positions)
fig,ax=plt.subplots()
ax.bar(bar_positions,bar_heights,0.3) # 以0.3 为宽度
plt.show()
```

ax.barh() # 可以实现横向画图
